```python
noise_list = ['is','a','this','the','...',' ','.']

def remove_noisedata(input_text):
  input_list = input_text.split()
  real_data = [word for word in input_list if word not in noise_list]
  data = " ".join(real_data)
  return data

print(remove_noisedata("this is the best fruit    ."))
```

```python
import re

def find_hashWord(text):
  regex = r"#(\w+)"
  matching_words = re.findall(regex,text)
  for word in matching_words:
    print(word)

if __name__ == "__main__":
  print("the words with hashtag are: ")
  find_hashWord("beautiful day #rainy #me #peace")
```

```python
text_columns = data.select_dtypes(include=['object'])
text_columns
```

```python
req_data = text_columns.loc[:,['user_description','text']]
req_data
```

```python
import re

regex = r"#\w+"

def getHashWord(data1):
      return re.sub(regex,'',data1)

for col in req_data.columns:
  req_data[col] = req_data[col].apply(getHashWord)

req_data
```

```python
data[req_data.columns] = req_data
data
```
